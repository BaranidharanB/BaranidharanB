```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the image 
img = cv2.imread("/Users/baranidharan/Desktop/DATA/sammy_face.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

# Applying Canny Edge Detector

```python
edge = cv2.Canny(image = img,threshold1 = 127 , threshold2 = 127  )
plt.imshow(edge)
```

```python
# varrying the threshold values
edge = cv2.Canny(image = img,threshold1 = 0 , threshold2 = 255  )
plt.imshow(edge)
```

# Formula for choosing the particular Threshold

```python
#1st Calculate the median value of the pixel
#2nd Setting the Lower and Upper Bound of the threshold
med_val = np.median(img)
med_val
```

```python
#2nd Setting the Lower and Upper Bound of the threshold
#0.7 ====> 70% of the median value
lower = int (max(0,0.7*med_val))
# upper threshold to either 130% of median value or the max 255, whichever is smaller
upper = int (max(255,1.3*med_val))
```

```python
edge = cv2.Canny(image = img,threshold1 = lower , threshold2 = upper   )
plt.imshow(edge)
```

```python
edge = cv2.Canny(image = img,threshold1 = lower , threshold2 = upper +100  )
plt.imshow(edge)
```
