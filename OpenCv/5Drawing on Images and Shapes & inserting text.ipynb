```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank_img.shape
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(0,360),pt2=(510,150),color=(255,0,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(0,360),pt2=(150,512),color=(255,0,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(0,360),pt2=(512,150),color=(255,0,0),thickness=10)
# drawing in the center
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,255,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

# Drawing a Circle

```python
#Drawing a circle
cv2.circle(blank_img,center=(100,100),radius=60,color=(0,0,255),thickness=10)
plt.imshow(blank_img)
```

```python
#Drawing a circle
cv2.circle(blank_img,center=(400,400),radius=60,color=(0,0,255),thickness=-1)
plt.imshow(blank_img)
```

# Drawing a line

```python
#Drawing a circle
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=6)
plt.imshow(blank_img)
```

# Inserting a text 

```python
#org refers to the origin
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text="hello",org=(10,500),fontFace=font,fontScale=5,color=(255,255,255),thickness=3,lineType = cv2.LINE_AA)
plt.imshow(blank_img)
```

# Drawing a Polygen

```python
img = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(img)
```

```python
#assigning the vertices of the polygen wrt [X,Y]
vertices = np.array([[100,300],[200,300],[400,300],[200,400]],dtype = np.int32)
```

```python
vertices
```

```python
vertices.shape
```

```python
#the opencv needs 3D .. so we need to convert and reshape it
pts = vertices.reshape((-1,1,2))
pts.shape
```

```python
pts
```

```python
#we should pass the image along with the pts list made above
cv2.polylines(img,[pts],isClosed = True, color = (0,255,0),thickness=5)
plt.imshow(img)
```
