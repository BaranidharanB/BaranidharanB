```python
image = cv2.imread("/Users/baranidharan/Desktop/DATA/dog_backpack.jpg")
image1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(image1)
```

```python
#Drawing a triangle
vertices = np.array([[250,700],[425,400],[600,700]],dtype = np.int32)
pts = vertices.reshape((-1,1,2))
pts.shape
```

```python
cv2.polylines(image1,[pts],isClosed = True, color = (255,0,0),thickness=5)
plt.imshow(image1)
```

```python
# FILL POLY - USED TO FILL THE COLOR INSIDE THE TRIANGLE
image1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.fillPoly(image1,[pts],color = (255,0,0))
plt.imshow(image1)
```
